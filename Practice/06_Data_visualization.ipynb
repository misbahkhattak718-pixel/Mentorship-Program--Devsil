***Data visualization***

```python
#import librariers:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the titanic data from seaborn librarier:
df = sns.load_dataset('titanic')
df
```

```python
df.head()
```

```python
df.plot(kind='scatter',x='age',y='fare',color='red')
```

```python
df.head()
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar')
```

```python
df.groupby('sex')['fare'].mean().plot(kind='bar')
plt.ylabel('Average Fare')
plt.xlabel('Gender')
```

```python
sns.barplot(data=df, x='survived',y= 'fare')
```

```python
sns.barplot(data= df,x='class',y='fare')
```

```python
sns.barplot(data= df,x='class',y='survived')
```

```python
sns.barplot(data=df,x='sex',y='fare',hue='class')
```

```python
sns.boxplot(data=df,x='sex',y='fare',hue='class')
```

```python
sns.scatterplot(data=df,x='sex',y='fare',hue='class')
```

```python
sns.scatterplot(data=df,x='age',y='fare',hue='class')
```

```python
sns.scatterplot(data=df,x='age',y='fare',hue='sex')
```

```python
sns.scatterplot(data=df,x='sex',y='class',hue='fare')
```
