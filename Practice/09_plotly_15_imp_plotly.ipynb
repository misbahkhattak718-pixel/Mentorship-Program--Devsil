***Most important plots:***

```python
import pandas as pandas
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.info()
```

```python
#scatter plots

fig1 = px.scatter(df,x='sepal_width',y='petal_length',color='species')
fig1.show()
```

```python
df.columns
```

```python
#Line plot:
fig2 = px.line(df,x='sepal_width',y='petal_length',color='species')
fig2.show()
```

```python
#Bar plot:
fig3 = px.bar(df,x='sepal_width',y='petal_length',color='species')
fig3.show()
```

```python
#Box plot:
fig4 = px.box(df,x='species',y='sepal_width',color='species')
fig4.show()
```

```python
#violin plot:
fig5 = px.violin(df,x='species',y='sepal_width',color='species')
fig5.show()
```

```python
#histogram:
fig6 = px.histogram(df,x='species',y='sepal_width',color='species')
fig6.show()
```

```python
# 7 pei chart:

fig7 = px.pie(df,values='sepal_width',names='species')
fig7.show()
```

```python
#8 scatter 3D plot:
fig8 = px.scatter_3d(df,x='species',y='sepal_width',z='petal_width',color='species')
fig8.show()
```

```python
# 9 Area chart:
df_area = df.sort_values(by=['species'])
fig9 = px.area(df_area,x='sepal_length',y='sepal_width',color='species')
fig9.show()
```

```python
# 10 sunburst plot:
df_sunburst=df.groupby(['species','sepal_width']).size().reset_index(name='counts')
fig10 = px.sunburst(df_sunburst,path=['species','sepal_width'],values='counts')
fig10.show()
```

```python
# 11 density_contour plot:
fig11 = px.density_contour(df,x='sepal_length',y='sepal_width',color='species')
fig11.show()
```

```python
# ternary plot:
fig12 = px.scatter_ternary(df,a='sepal_length',b='sepal_width',c='petal_width',color='species')
fig12.show()
```
