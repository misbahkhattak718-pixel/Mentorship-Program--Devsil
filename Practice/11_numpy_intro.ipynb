***NUMPY INTRODUCYION:***


```python
#CREATE ARRAY:
import numpy as np
a = np.arange(7)
print(a)
a.shape
```

```python
a2 = a[np.newaxis, :]
print(a2)
a2.shape
```

```python
a2 = a[: ,np.newaxis]
print(a2)
a2.shape
```

```python
a3= a2[: ,np.newaxis]
print(a3)
a3.shape
```
