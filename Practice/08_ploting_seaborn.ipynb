```python
# import librarier:
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
#Data:
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.info()
```

```python
sns.scatterplot(data=df,x='sex',y='fare',hue='age')
```

```python
sns.scatterplot(data=df,x='age',y='fare',hue='sex')
```

```python
name=['Misbah','noman','qasim']
height=[5.2,5.7,5.1]
#create the line chart:
fig=px.line(x=name,y=height,title="Height of student")
fig.show()
```

```python
name=['Misbah','noman','qasim']
height=[5.2,5.7,5.1]
#create the line chart:
fig=px.bar(x=name,y=height,title="Height of student")
fig.show()
```

```python
#plots on titanic data:
#scatter plots;
fig=px.scatter(df,x='age',y='fare',color='sex',title="plots of titanic data")
fig.show()
```

```python
import plotly.graph_objects as go

#heatmap plot is create:
data= np.random.rand(10,10)
heatmap= go.Heatmap(z=data)
fig=go.Figure(data=[heatmap])
fig.show()
```

```python
fig=px.imshow(df[['age','fare']].corr())
fig.show()
```

```python
# sunburst plot in titanic dataset:
fig = px.sunburst(df,path=['sex','class', 'who','alive', 'alone'],values='survived',color='sex')
fig.show()
```
