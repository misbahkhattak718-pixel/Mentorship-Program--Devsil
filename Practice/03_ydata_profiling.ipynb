```python
# Convert the whole Size column to bytes:

def convert_size(size):
    if isinstance (size, str):
        if 'M' in size:
            return float(size.replace('M', '')) * 1024
        elif 'k' in size:
            return float(size.replace('k', '')) *1024*1024
        elif  'Varies with device' is size: 
            return np.nan
```
