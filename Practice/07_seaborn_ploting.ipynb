***SEABORN AND PLOTING***

```python
#import librarier:
import pandas as pd
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data:
df=sns.load_dataset('titanic')
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df,x='age',y='fare',hue='sex')
```

```python
plt.figure(figsize=(12,8))
palette={'male':'blue','female':'red'}
sns.scatterplot(data=df,x='age',y='fare',hue='sex',palette=palette)
```

```python
plt.figure(figsize=(12,8))
palette={'male':'blue','female':'red'}
sns.scatterplot(data=df,x='age',y='fare',hue='sex',palette=palette)
#labels:
plt.xlabel('sex')
plt.ylabel('fare')
plt.title('scatter plot of Age vs Fare')
```

```python
plt.figure(figsize=(12,8))
palette={'male':'blue','female':'red'}
sns.scatterplot(data=df,x='age',y='fare',hue='sex',palette=palette)
#labels:
plt.xlabel('sex')
plt.ylabel('fare')
plt.title('scatter plot of Age vs Fare')
#saveing a plot 300dpi:
plt.savefig('scatter_plot.png',dpi=300)
```

```python


#style changes:
sns.set_theme(style='darkgrid',font_scale=2)
plt.figure(figsize=(12,8))
palette={'male':'blue','female':'red'}
sns.scatterplot(data=df,x='age',y='fare',hue='sex',palette=palette)
#labels:
plt.xlabel('sex')
plt.ylabel('fare')
plt.title('scatter plot of Age vs Fare')
#saveing a plot 300dpi:
plt.savefig('scatter_plot.png',dpi=300)
```

***BOX PLOT:***

```python
#style changes:
sns.set_theme(style='darkgrid',font_scale=2)
plt.figure(figsize=(12,8))
palette={'First':'green','Second':'red','Third':'blue'}
sns.boxplot(data=df,x='sex',y='age',hue='class',palette=palette)
#labels:
plt.xlabel('sex')
plt.ylabel('class')
plt.title('scatter plot of Age vs Fare')
#saveing a plot 300dpi:
plt.savefig('scatter_plot.png',dpi=300)
```

```python
#style changes:
sns.set_theme(style='darkgrid',font_scale=2)
plt.figure(figsize=(12,8))
palette={'First':'green','Second':'red','Third':'blue'}
sns.boxenplot(data=df,x='sex',y='age',hue='class',palette=palette)
#labels:
plt.xlabel('sex')
plt.ylabel('class')
plt.title('scatter plot of Sex vs Class')
#saveing a plot 300dpi:
plt.savefig('scatter_plot.png',dpi=300)
```

```python
#style changes:
sns.set_theme(style='darkgrid',font_scale=2)
plt.figure(figsize=(12,8))
palette={'First':'green','Second':'red','Third':'blue'}
sns.barplot(data=df,x='sex',y='age',hue='class',palette=palette)
#labels:
plt.xlabel('sex')
plt.ylabel('class')
plt.title('scatter plot of Sex vs Class')
#saveing a plot 300dpi:
plt.savefig('Bar_plot.png',dpi=300)
```

```python
#style changes:
sns.set_theme(style='darkgrid',font_scale=2)
plt.figure(figsize=(12,8))
palette={'First':'green','Second':'red','Third':'blue'}
sns.violinplot(data=df,x='sex',y='age',hue='class',palette=palette)
#labels:
plt.xlabel('sex')
plt.ylabel('class')
plt.title('scatter plot of Sex vs Class')
#saveing a plot 300dpi:
plt.savefig('Bar_plot.png',dpi=300)
```
